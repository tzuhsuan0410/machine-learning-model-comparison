```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.preprocessing import MinMaxScaler
```

# 3. Data scaling with minimax technique

```python
#load data(sometimes need to use "/")
dummied_new_df = pd.read_csv('C:/Users/user/Desktop/Final_assignment/Data/organics_deleteNA_dummies.csv')
dummied_new_df
```

```python
#minimax scaling
MMscaler=MinMaxScaler(feature_range=(0, 1))
scaling=MMscaler.fit_transform(dummied_new_df)
scaled_data=pd.DataFrame(data=scaling)
scaled_data.columns=['X4','X7','X8','X9','X1_1','X1_2','X2_1','X2_2','X2_3','X2_4','X2_5','X3_1','X3_2','X3_3','X3_4','X5_1','X5_2','X5_3','X5_4','X5_5','X5_6','X5_7','X6_1','X6_2','X6_3','X6_4','X6_5','X6_6','X6_7','X6_8','X6_9','X6_10','X6_11','X6_12','X10_0','X10_1','Y']
scaled_data.head()
```

# 4. Set dependent variable(Y or target) & independent variable(X)

```python
#set x (make prediction) with minimax
scaled_x=scaled_data.drop(['Y'],axis=1).copy()
#set y (want to predict)
y=dummied_new_df['Y'].copy()
```

# 5.  Hyper parameter tuning

```python
#Split the data (split into 80% training data & 20% testing data) (lock seed) 
#(stratify:make sure both training & testing data set contains same percentage of result)
MAE_all_list=[]
MAPE_all_list=[]
MSE_all_list=[]
RMSE_all_list=[]
for i in range(1,11):
    x_train,x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.2,random_state=i)

    # Initializing Classifiers
    clf1 = Lasso(random_state=4,max_iter=10000)

    # Setting up the parameter grids
    alphas=np.logspace(-1, -0.001, num=30)
    param_grid1 = [{'alpha':alphas}]
    cross_val = KFold(n_splits=5,shuffle=True,random_state=4)

    gridS_model = GridSearchCV(estimator=clf1,
                                    param_grid=param_grid1,
                                    scoring='neg_root_mean_squared_error',
                                    n_jobs=-1,
                                    cv=cross_val,
                                    verbose=0,
                                    refit=True)

    
    gridS_model.fit(x_train, y_train) 
    training_score=gridS_model.best_score_  
    best_parameter=gridS_model.best_params_
    train_model=gridS_model.best_estimator_\
    
    #print training result
    print('Round',i)
    print('random_state=',i)
    print('Training score:',training_score)
    print('Best Hyper-parameter set:',best_parameter)
    
    #model result set square = False to return RMSE
    y_pred=gridS_model.predict(x_test)
    MAE=mean_absolute_error(y_test,y_pred)
    MAPE=mean_absolute_percentage_error(y_test,y_pred)
    MSE=mean_squared_error(y_test,y_pred,squared=True)
    RMSE=mean_squared_error(y_test,y_pred,squared=False)


    print('---------------------------------------')
    #print('r2:    ',round(R2,4))
    #print('adj r2:',round(adj_R2,4))
    #print('---------------------------------------')
    print('Metrics')
    print('MAE:   ',round(MAE,2))
    print('MAPE:  ',round(MAPE,2))
    print('MSE:   ',round(MSE,2))
    print('RMSE:  ',round(RMSE,2))
    print('---------------------------------------')
    print('---------------------------------------\n\n')
    
    MAE_all=[round(MAE,2)]
    for r in MAE_all:
        MAE_all_list+=[r]
        
    MAPE_all=[round(MAPE,2)]
    for r in MAPE_all:
        MAPE_all_list+=[r]
        
    MSE_all=[round(MSE,2)]
    for r in MSE_all:
        MSE_all_list+=[r]
    
    RMSE_all=[round(RMSE,2)]
    for r in RMSE_all:
        RMSE_all_list+=[r]
            


best_RMSE=min(RMSE_all_list)
best_round=RMSE_all_list.index(best_RMSE)+1
print('Best_RMSE:',best_RMSE)
print('Best Round:',best_round)
print('---------------------------------------')
print('MAE_mean:',round(np.mean(MAE_all_list),2),',MAE_std:',round(np.std(MAE_all_list),2))
print('MAPE:',round(np.mean(MAPE_all_list),2),',MAPE_std:',round(np.std(MAPE_all_list),2))
print('MSE_mean:',round(np.mean(MSE_all_list),2),',MSE_std:',round(np.std(MSE_all_list),2))
print('RMSE_mean:',round(np.mean(RMSE_all_list),2),',RMSE_std:',round(np.std(RMSE_all_list),2))
# MAE_all_list
# MAPE_all_list
# MSE_all_list
# RMSE_all_list
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.2,random_state=2)
model = Lasso(alpha=0.9977000638225533)
model.fit(x_train,y_train)
```

```python
importance=model.coef_
importance_sort = np.argsort(importance)[::-1]
df_col = dummied_new_df.columns[0:]
print('Feature Importance')
print('----------------------------------------------')
for i,v in enumerate(importance):
    df_columns=df_col[i]
    print('%s \t  Feature_Imp= %.5f' % (df_columns,v))
print('\n')    
print('Feature ranking')
print('----------------------------------------------')
    
# 依序迭代出重要特徵
for f in range(x_train.shape[1]):
    print(f"{f+1:>2d}) {df_col[importance_sort[f]]:<30s} {importance[importance_sort[f]]:.3f}")
```
